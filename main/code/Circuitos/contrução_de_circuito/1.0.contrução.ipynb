```python
from qiskit import *
from qiskit.circuit.library import HGate
```

```python
circuit = QuantumCircuit(2)
circuit.append(
    HGate(),
    [0]
)
```

```python
circuit.draw('mpl')
```

### Combinação de circuitos com método compose

```python
circuit_1 = QuantumCircuit(4, name="circuit_1")
circuit_1.x(0)

circuit_2 = QuantumCircuit(2, name='circuito_2')
circuit_2.y(0)
circuit_2.z(1)

#Combinação dos qubits[circuit_1](0,1)--> qubits[circuit_2](1,3) 

combined = circuit_1.compose(circuit_2, qubits=[1,3])
combined.draw("mpl")
```

### Circuito como instrução

```python
circuit_3 = QuantumCircuit(4, name="circuit_3")
circuit_1.x(0)

circuit_4 = QuantumCircuit(2, name='circuito_4')

inst = circuit_4.to_instruction()
circuit_3.append(inst, [1,3])
circuit_3.draw("mpl")
```

```python
gate = circuit_4.to_gate().control()
circuit_3.append(gate, [0 ,1 ,3])
circuit_3.draw("mpl")
```

### Decomposição:

```python
circuit_5 = QuantumCircuit(4, name="circuit_3")
circuit_5.append(gate, [0,1,3])
circuit_5.draw('mpl')
circuit_5.decompose().draw("mpl")
```
