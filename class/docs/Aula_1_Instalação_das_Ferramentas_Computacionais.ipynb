# Instalando o Anaconda 

## Windows

1. Acesse o site do Anaconda para fazer o Download segundo a versão do seu Windows:

https://www.anaconda.com/download/success

2. Após o download abra o arquivo .exe baixado, recomendamos o uso opções de instalação padrão. 


## Linux (Ubuntu)

1. Comece atualizando o gerenciador de pacotes local. Abra uma janela de terminal e digite o seguinte:

```sudo apt-get update```

2. Acesse o site do Anaconda para fazer o Download segundo a versão do seu Linux:

https://www.anaconda.com/download/success

3. Abra o seu terminal na pasta onde você efetuou o download e altere as permissões do arquivo .sh baixado:

```chmod +x Anaconda3-2024.06-1-Linux-x86_64.sh```

```chmod``` é o comando para alterar permissões.
```+x``` adiciona a permissão de execução (execute) para o arquivo.

4. Execute o arquivo no terminal para iniciar o processo de instalação:

```./Anaconda3-2024.06-1-Linux-x86_64.sh```

Você deve ver o seguinte output:

```Welcome to Anaconda3 2024.06```

```In order to continue the installation process, please review the license agreement.```

```Please, press ENTER to continue```
```>>> ```

5. Pressione ENTER para continuar. Para percorrer o termo de licença, use sempre a tecla ENTER. Depois de revisar o termo, você será solicitado a aprovar os termos da licença:

```Do you approve the license terms? [yes|no]```

Digite ```yes``` para aceitar a licença.

6. Escolha o local de instalação:

```Anaconda3 will now be installed into this location:```

```/home/linuxize/anaconda3```

    ```- Press ENTER to confirm the location```
    
    ```- Press CTRL-C to abort the installation```
    
    ```- Or specify a different location below```
    
Recomendamos o uso opções de instalação padrão. Pressione ENTER para confirmar a localização Padrão.

7. A instalação pode levar algum tempo, para ela ser concluída, o script perguntará se você deseja executar ```conda init```. Digite ```yes```.

Isso é importante pois adicionará a ferramenta de linha de comando ```conda``` ao ```PATH``` do seu sistema.

# Primeiros Passos

## Criando Circuitos Quânticos

O Primeiro passo é sempre importar o Qiskit


````python
from qiskit import *
````

````python
Qr = QuantumRegister(1, 'q') # Registrando 1 bit quântico no meu Circuito
````

````python
Cr = ClassicalRegister(1, 'c') # Registrando 1 bit clássico no meu Circuito
````

````python
circuit = QuantumCircuit(Qr, Cr) #criando o circuito
circuit.reset(Qr[0]) #resetar esse qubit
````

````python
%matplotlib inline 
#premite plotar gráficos neste notebook
circuit.draw()
````

````python
circuit.draw(output = 'mpl')
````

# Vizualizando na Esfera de Bloch

````python
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
````

````python
state = Statevector(circuit)
plot_bloch_multivector(state)
````

````python
circuit.h(Qr[0])
````

````python
circuit.draw(output = 'mpl')
````

````python
state = Statevector(circuit)
plot_bloch_multivector(state)
````

````python
circuit.z(Qr[0])
````

````python
circuit.draw(output = 'mpl')
````

````python
state = Statevector(circuit)
plot_bloch_multivector(state)
````

## Portas Lógicas Quânticas

````python
# Definindo um novo circuito
circuit2 = QuantumCircuit(2)
circuit2.h(0)
circuit2.x(1)
 
circuit2.draw(output = 'mpl')
````

````python
from qiskit.quantum_info import DensityMatrix

matrix = DensityMatrix(circuit2)
plot_bloch_multivector(matrix, title='Esfera de Bloch', reverse_bits=False)
````
