```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
f = np.sin(x)
dx = x[1] - x[0]
df = np.gradient(f, dx)  # aproximação da derivada

plt.plot(x, f, label='f(x) = sin(x)')
plt.plot(x, df, label="f'(x) ≈ cos(x)", linestyle='--')
plt.legend()
plt.grid()
plt.show()
```
